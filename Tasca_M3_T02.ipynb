* **SecurityDelay**     : Retraso por seguridad, en minutos. Aquín se incluyo todo tipo de poblema originado por fallas en la seguridad.
* **LateAircraftDelay** : Retraso de un vuelo, en minutos. Se debe a que el avión utilizado para el vuelo en cuestión tuvo un retraso en llegar al aeropuerto, por diversas razones.

**3. Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).**

```python
# Ahora crearemos una nueva columna que nos muestre la velocidad mediana del vuelo.

"""Para este ejercicio necesitamos utilizar columnas que nos muestren la distancia por un lado y por el otro el tiempo que
le toma a un vuelo recorrer dicha distancia. Para la distancia solo tenemos el campo 'distance' como referencia por lo que no
necesitamos pensarlo mucho. Mientras tanto para el tiempo de recorrido tenemos dos campos 'ActualElapsedTime' y 'AirTime'. 
Como se nos solicita la velocidad media del vuelo podría no servirnos de mucho considerar todo aquel tiempo donde el avión
no se encuentra en el aire a muy bajas velocidades, es decir, el TaxiIn y TaxiOut. Eso podría "ensuciar" el resultado final,
por tanto considero que bajo este criterio el dato de tiempo de vuelo en el aire es mucho más preciso para estos efectos. De
esta manera utilizaremos 'AirTime'."""

""" Para seguir un paso a paso más pedagógico y ordenado para obtener el resultado solicitado, crearé columnas nuevas con 
la unidad de medida en km y horas antes de obtener la velocidad media del vuelo """

sample_df['dist_km'] = sample_df['Distance']* 1.60934
sample_df['Air_hr'] = sample_df['AirTime']/ 60
sample_df['Vel_mean'] = sample_df['dist_km']/sample_df['Air_hr']

sample_df.head()
```

```python
# En esta oportunidad crearemos una columna donde podamos ver si el vuelo se ha retrasado o no.

""" Aquí la columna a utilizar será 'ArrDelay' donde ya sabemos que para que un vuelo se cuente como retrasado, debe tener
más de 15 min de retraso. Por lo tanto crearemos la columna 'Dlyed_no_Dlyed' donde utilizaremos la función `where` de NumPy donde expresaremos
que todo aquel registro que tenga un valor superior a 15 min (> 15) será re-categorizado en el nuevo campo como dlayed y si lo anterior no sucede,
es decir que el retraso sea < 15 min, será categorizado como 'on_time'."""

sample_df['dlyed_no_dlyed'] = np.where(sample_df['ArrDelay'] > 15,'dlayed','on_time')
print(sample_df[['ArrDelay','dlyed_no_dlyed']])
```

```python
# ahora crearemos una nueva columna que nos podría servir para algún exercicio de más adelante si quisiésemos utilizarla.
# Crearemos la columna "Rutas"

""" Lo que haremos será concatenar mediante "cat" el origen (Origin) y el destino (Dest). Para asegurarnos de que "cat" funcione correctamente
utilizamos la conversion a cadena de texto (str). Luego, dentro de la función función cat asignamos como primer argumento el final de la nueva cadena
"Dest" y como segundo argumento "sep" (separador) la forma en la que uniremos ambas cadenas de texto, es decir a través de un espacio vacío " ". """


sample_df['Rutas']=sample_df.Origin.str.cat(sample_df.Dest, sep=" ")
sample_df['Rutas']
```

**4. Fes una taula de les aerolínies amb més endarreriments acumulats**.

```python
# Probaremos dos maneras de realizar el cálculo solicitado
# La primera será mediante groupby

""" Como nos preguntan sobre atrasos entonces debemos en primer lugar identificar dónde se encuentran. Para eso farem servir la columna
ya creada 'dlyed_no_dlyed' == "dlayed". De esta manera ya los tenemos identificados para su posterior conteo. Luego agruparemos dicha información
por el campo 'UniqueCarrier' para finalmente realizar el conteo. Finalmente imprimimos la variable ordenada de manera ascendente """

aero_dly_acum=sample_df[sample_df.dlyed_no_dlyed=="dlayed"].groupby(["UniqueCarrier"])["dlyed_no_dlyed"].value_counts()
print(aero_dly_acum.sort_values(ascending=False))
```

```python
# Ahora probaremos la segunda forma de hacerlo.
# primero creamos una variable para tener solo los vuelos atrasados desde la columna ya creada y luego realizamos el conteo para cada código de operador (Carrier)

dlayed = sample_df[sample_df['dlyed_no_dlyed'] == 'dlayed']
aero_dlayed = dlayed['UniqueCarrier'].value_counts()

print(aero_dlayed)
```

**5. Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.**

5.1 Quins són els vols més llargs? 

```python
""" Para este caso utilizaremos la columna creada 'dist_km' para expresar el resultado en dicha unidad de medida. 
Además llamaremos a la función 'sort_values" para ordenar la columna de manera ascendente mediante false. Finalmente para mostrar
los primeros 30 vuelos más largos utilizaremos el argumento n=30 en head para su visualización.'""" 

# He puesto el valor 30 porque de esta manera es posible visualizar distancias diferentes. 

Vuelos_mas_largos = sample_df['dist_km'].sort_values(ascending = False).head(30)
print("Los 30 vuelos más largos en (km) son: ")
print(Vuelos_mas_largos)
```

5.2 I els més endarrerits?

```python
""" Como el elay en la llegada esta calvulado en minutos y se nos solicitan los más atrasados, se presupone que la lista 
arrojará números muy grandes. por esta razón se creará una variable que almacene los valores en horas, de manera tal que al momento
de generar la tabla la información sea mucho más legible. El resto de la operación ya se realizó en el ejercicio anterior.""" 

arr_delay_hr = sample_df['ArrDelay'] / 60
Vuelos_mas_atrasados = arr_delay_hr.sort_values(ascending = False).head(10)

print("Los 10 vuelos más atrasados (en hr): ")
print(Vuelos_mas_atrasados)
```

5.3 Busca les rutes més llargues

```python
""" El siguiente ejercicio esta hecho en una línea de código pero en realidad son 3 partes. La primera parte es aquella que finaliza
con el groupby. Este método lo utilizamos para generar todas las combinaciones posibles de rutas para todos los vuelos a través de
su origine y su destino. En la egunda parte convocamos a la función .agg, la cual nos sirve para calcular estadísticas descriptivas
en este caso para cada ruta de vuelo. Como necesitamos la ruta más larga entonces utilizamos .max y se lo aplicamos a la columna
que nos sirve para obtener el cálculo, es decir, dist_km la cual nos entrega la distancia (para este caso la distancia de la ruta).
Finalmente nos queda ordenar la tabla mediante sort_values tomando como referencia la columna distancia (función utilizada previamente). 
Para acabar solicitamos que nos arroje solo las 10 rutas más largas del dataset. 
"""
# Me hubiese gustado quitar los duplicados de las rutas eliminando ya sea la ida o la vuelta, pero no logré entender cómo hacerlo.

rutas_mas_largas = sample_df.groupby(['Origin','Dest']).agg({'dist_km':'max'}).sort_values(by='dist_km', ascending=False).head(10)
rutas_mas_largas
```

```python
# Otra forma de hacerlo utilizando nuestra columna 'Rutas', creada previamente.

rutes_mas_largas = sample_df.groupby('Rutas').agg({'dist_km':'max'}).sort_values(by='dist_km', ascending=False).head(10)
rutes_mas_largas
```

5.4 Rutes que acumulen més retards

```python
# Primera forma para lograr un resultado.
# he pensado esta primera manera como aproximación a lo solicitado puesto que no he podido llegar a una manera concreta.

""" Mediante un groupby realizaré la agrupación de los tiempos de retraso enrelación a las rutas ya creadas previamente. Calcularé
el máximo valor en los tiempos de estas rutas y las ordenadré de manera ascendente. Entonces tenemos el máximo tiempo en min de retraso
por rutas"""

rutas_adly = sample_df.groupby('Rutas')['ArrDelay'].max()
rutas_adly.sort_values(ascending=False)
```

```python
# Esta segunda forma no considera tiempo en minutos sino valores absolutos acumulados de retraso para la llegada de los vuelos agrupados en rutas. 
# Finalmente tenemos el conteo de la cantidad de vuelos retrasados en arribo por rutas, ordenados de mayor a menor (ascendente)

rutas_dly_acum=sample_df[sample_df.dlyed_no_dlyed=="dlayed"].groupby(["Rutas"])["dlyed_no_dlyed"].value_counts()
print(rutas_dly_acum.sort_values(ascending=False))
```
